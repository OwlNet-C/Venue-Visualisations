```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
age = pd.read_csv("VenueAge.csv",index_col=0)
spending = pd.read_csv("VenueSpend.csv",index_col=0)
daily_visitors = pd.read_csv("VenueDailyVisitors.csv",index_col=0)
distance = pd.read_csv("VenueDistance.csv", index_col=0)
gender = pd.read_csv("VenueGender.csv",index_col=0)
duration = pd.read_csv("VenueDuration.csv", index_col=0)
```

```python
data = pd.DataFrame(round(daily_visitors.mean()), columns = ['Mean visitors per day'], index=distance.index)
#Defining how data looks through csv files
```

```python
data['Avg spend (Â£)'] = spending.values
data['Distance (Miles)'] = distance.values
data['Gender (Female vs Male Percentage)'] = gender.values
data['Duration (Mins)'] = duration.values
data['Age (Years)'] = age.values
```

```python
plt.figure(figsize=(8, 8))
corr = data.corr()
ax = sns.heatmap(corr, vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(220, 20, n=200),
    square=True)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45,
    horizontalalignment='right')
plt.show()
```
